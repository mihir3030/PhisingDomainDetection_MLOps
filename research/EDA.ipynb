```python
#checking outlier using boxplot
fig, ax = plt.subplots(ncols=7, nrows=4, figsize=(25,20))
index = 0
ax = ax.flatten()

for col, value in df_second.items():
    sns.boxplot(y=col, data=df_second, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

there are many outliers in columns there are serval methos we can use. like StandardScaler and LogTransformation etc..

```python
# checking distribution of dataset
fig, ax = plt.subplots(ncols=7, nrows=4, figsize=(25,20))
index=0
ax = ax.flatten()

for col, value in df_second.items():
    sns.histplot(value, ax=ax[index], kde=True)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
